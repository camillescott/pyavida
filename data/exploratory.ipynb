```python
dfs = []
for exp in ['A', 'B']:
    for cond in range(4):
        for rep in range(10):
            fn = 'lambda_2014-02-22-p_genarch_{}_{}_{}.dat'.format(exp, cond, rep)
            try:
                df = pd.read_csv(fn)
                df['experiment'] = exp
                df['condition'] = cond
                df['replicate'] = rep
                dfs.append(df)
            except IOError as e:
                print e
```

```python
data = pd.concat(dfs)
```

```python
data['g_ind'] = data.u_birth // 100
```

```python
data_grouped = data.loc[:,['Lambda','u_birth', 'fitness', 'experiment', 'condition', 'g_ind']].groupby(['experiment','condition','g_ind'])
```
