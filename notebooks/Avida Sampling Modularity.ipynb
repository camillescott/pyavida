```python
# sample 2 instructions N times from the coding regions, return N and the number of times
# the samples were from the same contiguous run of instructions
#
# in other words, we are measuring the probablitiy that two randomly sampled coding positions
# are in the same module
#
# return genome length, number coding insts, and number of samples landing in same coding region

def check_region(G, x, y, t=1):
    for i in xrange(x,y):
        if G[i] != t:
            return False
    return True

def P_m(G, N=1000, t=1, S=2):
    results = np.zeros(N)
    coding = np.where(G >= t)[0]
    #print coding
    if len(coding) >= S:
        for i in xrange(N):
            N_s = 0.0
            samples = np.sort(np.random.choice(coding, size=S, replace=False))
            for j in xrange(1,S):
                if check_region(G, samples[j-1], samples[j]):
                    N_s += 1.0
            results[i] = N_s / (S-1)
    
    results += np.random.normal(loc=0.0, scale=.00001, size=N)
    results = np.clip(results, 0.0, 1.0)
    
    return np.array([np.mean(results), np.var(results)])
```

```python
def random_genome(L, N_c):
    '''
    Generate a random (binary) array of length L with N_c coding instructions
    '''
    
    coding = np.random.randint(0, high=L, size=N_c)
    G = np.zeros(L)
    np.put(G, coding, 1)
    return G


def random_modular_genome(L, N_c, M=.25):
    '''
    Generate a random genome with P_m > .25
    WARNING: This function is terrible and takes a long time to run,
        because modular genomes are exceedingly rare
    '''
    
    M_c=0
    i=0
    while(M_c < M):
        G = random_genome(L, N_c)
        i += 1
        M_c = P_m(G, N=100)[0]
    return G, i


def random_pop(N=1000, L=100, N_c=25):
    '''
    Generate a random population of size N.
    L can be either an int for equal length genomes, or a numpy array of lengths
    N_c has the same restrictions as L    
    '''
    
    pop = []
    if type(L) == int and type(N_c) == int:
        for i in xrange(N):
            pop.append(random_genome(L, N_c))
    elif type(L) in [np.ndarray, list] and type(N_c) in [np.ndarray, list]:
        if len(L) == len(N_c) and len(L) == N:
            for i in xrange(N):
                pop.append(random_genome(L[i], N_c[i]))
        else:
            raise IndexError('L and N_c must be lists of length N')
    else:
        raise TypeError('L and N_c must be ints or lists of length N')
    return pop


def calc_random(L, N_c, N=1000):
    '''
    Convenience function for returning P_m of a randomly generated genome
    with the given parameters.
    '''
    
    G = random_genome(L=L, N_c=N_c)
    return P_m(G, N=N)


def E(L, N_c, N=1000, Ns=1000):
    '''
    Get the expected P_m for a genome L, N_c. 
    Use N genomes, with Ns passed to the P_m function
    '''
    
    results = np.zeros((N,2))
    for i in xrange(N):
        Rs = calc_random(L=L, N_c=N_c, N=Ns)
        results[i,:] = Rs[:2]
    return results
```

Here, generate two distributions with the same parameters; show that they are, in fact, the same distribution, with a simple t-test. It appears to work!

```python
D3 = E(100, 25, N=1000, Ns=1000)
D4 = E(100, 25, N=1000, Ns=1000)
```

```python
t, p = sp.stats.ttest_ind(D3[:,0], D4[:,0])
print 't-test:', t, p

#print len(D3), 'samples'
#print np.min(D3[:,0])
sns.set_color_palette("hls")
sns.distplot(D3[:,0], kde_kws={'label':'Random 1'})
sns.distplot(D4[:,0], kde_kws={'label':'Random 2'})
axis(xmin=0)
```

The minimum and maximum-modularity genomes. As we would expect, we get distributions centered around 0 and 1. The slight shift is due to the "wobble" added in $P_m(g)$, which prevents returning junk distributions of all 0's or all 1's which break a lot of tools.

Real Data
---------

Now we show the system on actual data from a run. 

```python
dfs = []
for i in xrange(10):
    d = os.path.join(datadir, '2014-02-14_genarch_A_0_{i}'.format(i=i))
    fn = os.path.join(d, 'dom_lineage_orgs.dat')
    o_df = orgmap.get_org_mat(fn)
    dfs.append((o_df, d))
```

```python
def mod_func(X, d):
    org_id = int(X[1])
    fn = os.path.join(d,'dom_lineage/tasksites.org-{i}.dat'.format(i=org_id))
    #print fn
    _, org_mat = orgmap.parse_taskmap(fn)
    #print org_mat
    mod = P_m(org_mat)
    #print mod
    return mod[0]
```

```python
for i in xrange(10):
    o_df, d = dfs[i]
    P_m_D = o_df[o_df.u_birth > 0].apply(mod_func, axis=1, raw=True, args=(d,))
    o_df.insert(6, 'P_m', P_m_D)
```

```python
for o_df, d in dfs:
    X = o_df[o_df.u_birth > 0].u_birth
    Y = o_df[o_df.u_birth > 0].P_m
    plot(X, Y)
```
